### Joint plots

```python
from matplotlib import pyplot as plt    # visualization
import seaborn as sns    # statistical/advanced visualization

%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
j = plt.jointplot(data = movies, x = 'CriticRating', y = 'AudienceRating')
# Note the error
```

```python
j = sns.jointplot(data = movies, x = 'CriticRating', y = 'AudienceRating')
# Do shift + tab on top of jointplot. Note 'kind' shows the kinds of plots allowed
```

Interpretation - Note that CrtiticRating and Audience Rating have positive correlation

```python
j = sns.jointplot(data = movies, x = 'CriticRating', y = 'AudienceRating', kind = 'scatter')
# Note the outliers
```

```python
j = sns.jointplot(data = movies, x = 'CriticRating', y = 'AudienceRating', kind = 'hex')
```

```python
j = sns.jointplot(data = movies, x = 'CriticRating', y = 'AudienceRating', kind = 'kde')
# kde is kernel density estimation
```

### Subplots
- To get multiple graphs in the same figure

```python
f,axes = plt.subplots(1,2, figsize=(14,4))

fig1 = sns.kdeplot(movies.BudgetMillions, movies.AudienceRating, ax=axes[0])
fig1 = sns.kdeplot(movies.BudgetMillions, movies.CriticRating, ax=axes[1])
```

```python
f,axes = plt.subplots(1,3, figsize=(16,4))
```

```python
# Box Plots
b1 = sns.boxplot(data = movies, x='Genre', y='CriticRating')
# Observe which genre has the lowest critic rating: horror, romance
# Highest avg. CriticRating Thriller
# Min, 25% percentile line, mean, 75%line, highest
```

```python
# Violin plot
v1 = sns.violinplot(data = movies, x='Genre', y='CriticRating')
```

```python
b1 = sns.boxplot(data = movies[movies.Genre == 'Drama'], x='year', y='AudienceRating')
# Note the outliers in 2010
```

```python
# FacetGrid - to visualize multiple graphs
fig2 = sns.FacetGrid(movies, row = 'Genre', col = 'year', hue = 'Genre')
fig2 = fig2.map(plt.scatter, 'CriticRating', 'AudienceRating')   # To map scatter plots in FacetGrid
```

**How to visualize univariate, bivariate, multivariate at the same time?**
- Univariate: histplot
- Bivariate: scatter, lmplot
- Multivariate: facetgrid, subplot

- Anonymous functions/ Lambda functions
    - Filter
    - Map
    - Reduce

```python
# FacetGrid - to visualize multiple graphs
fig3 = sns.FacetGrid(movies, row = 'Genre', col = 'year', hue = 'Genre')
fig3 = fig3.map(plt.hist, 'BudgetMillions')   # To map histogram plots in FacetGrid
```
