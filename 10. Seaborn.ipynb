```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

%matplotlib inline
plt.rcParams['figure.figsize'] = 4,2

import warnings
warnings.filterwarnings('ignore') # os error
```

```python
df = pd.read_csv(r'C:\Users\sruja\datasets\GDP data.csv')
df
```

```python
df['InternetUsers']
```

#### Statistics - Univariate analysis

```python
vis1 = sns.distplot(df['InternetUsers'])
```

```python
vis2 = sns.distplot(df['InternetUsers'], bins = 10)
```

```python
vis3 = sns.displot(df['InternetUsers'])
```

#### Bivariate Analysis

```python
vis4 = sns.boxplot(data = df, x = 'IncomeGroup', y = 'BirthRate')
```

```python
vis5 = sns.lmplot(data = df, x = 'InternetUsers', y = 'BirthRate', fit_reg = False)
# lm: Linear Model
```

```python
vis6 = sns.lmplot(data = df, x = 'InternetUsers', y = 'BirthRate')
```

```python
vis7 = sns.lmplot(data = df,x = 'InternetUsers', y = 'BirthRate',
                  fit_reg = False,hue = 'IncomeGroup') #hue - parameter for color
```
